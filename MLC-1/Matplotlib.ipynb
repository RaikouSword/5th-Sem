```python
import matplotlib.pyplot as mat
import numpy as np

x = np.linspace(0,10,100)
y = np.sin(x)
mat.plot(x,y)
mat.title('Simple Line Plot')
mat.xlabel('X-axis')
mat.ylabel('Y-axis')
mat.show()
```

```python
import matplotlib.pyplot as mat
import numpy as np

x = np.linspace(-150,150,100)
y = x**2
mat.plot(x,y)
mat.title('Parabola')
mat.xlabel('X-axis')
mat.ylabel('Y-axis')
# print(y)
mat.show()
```

```python
date = ['25/12','26/12','27/12']
temp = [8.5,10.5,6.8]
mat.plot(date,temp)
mat.xlabel('Date')
mat.ylabel('Temp')
mat.title('Date wise Temp')
mat.grid(True)
mat.yticks(temp)
mat.show()
```

```python
y= np.sin(x)
y1 = np.cos(x)
mat.plot(x,y,label='sin(x)')
mat.plot(x,y1,label='cos(x)')
mat.title('Multiple Lines')
mat.xlabel('X')
mat.ylabel('Y')
mat.legend()
mat.show()
```

```python
x= np.linspace(0,10,100)
y = np.sin(x)
y1 = np.cos(x)
fig,axis = mat.subplots(2)
axis[0].plot(x,y,'x')
axis[0].set_title('sin(x)')
axis[1].plot(x,y1,'k')
axis[1].set_title('cos(x)')
mat.tight_layout()
mat.show()
```

```python
x_vals=['A','B','C','D']
y_vals = [3,7,5,4]
mat.bar(x_vals,y_vals,color='darkgreen')
mat.xlabel('Categories')
mat.ylabel('Values')
mat.title('Bar Plot Example')
mat.show()
```

```python
data = np.random.randn(1000)
mat.hist(data,bins=30,color='darkgreen',alpha=0.8)
mat.title('Histogram Example')
mat.xlabel('Value')
mat.ylabel('Frequency')
mat.show()
```

```python
x1 = np.random.normal(0,1,1000)
x2 = np.random.normal(-2,1,1000)
x3 = np.random.normal(3,2,1000)
mat.hist(x1,bins=10,color='beige',edgecolor='black')
mat.hist(x2,bins=10,color='teal',edgecolor='black')
mat.hist(x3,bins=10,color='gray',edgecolor='black')
mat.show()
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
colors = np.random.rand(50)
sizers =1000*np.random.rand(50)
mat.scatter(x,y,c=colors,s=sizers,alpha=0.8)
mat.title('Scatter Plot Example')
mat.xlabel('X')
mat.ylabel('Y')
mat.show()
```

```python
np.random.seed(42)
X = np.random.rand(50)
Y = np.random.rand(50)
mat.scatter(X,Y,color='red',marker='o',s=100,edgecolors='black',alpha=0.8)
#s = Sets the marker size
# edgecolor = Sets the color of the marker edges
# alpha =Adjusts the transparency of the markers
# mat.legend(loc=7)
mat.title('Scatter Plot')
mat.show()
```
