```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as mat
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
```

```python
#1
p1 =0.4
n_samples1 = 10000
bern1 = np.random.binomial(1,p1,size=n_samples1)
print('sample mean1: ', bern1.mean())
print('sample var1: ', bern1.var())


#2
fig,axes = mat.subplots(1,2,figsize=(12,8))
vals1,counts1 = np.unique(bern1, return_counts=True)
axes[0].bar(vals1,counts1/len(bern1))
axes[0].set_title('PMF1')
axes[0].set_xlabel('Value')
axes[0].set_ylabel('Probability')

#3
p2 = 0.3
n_samples2=2000
bern2 = np.random.binomial(1,p,size=n_samples)
vals2,counts2 = np.unique(bern2, return_counts=True)
print('sample mean2: ', bern2.mean())
print('sample var2: ', bern2.var())
axes[1].bar(vals2,counts2/len(bern2))
axes[1].set_title('PMF2')
axes[1].set_xlabel('Value')
axes[1].set_ylabel('Probability')
```
